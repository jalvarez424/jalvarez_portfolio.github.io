```python
# looking at different plt styles to use
print(plt.style.available)
```

```python
# use dark_background plt style
plt.style.use("dark_background")
```

```python
# let's use a heatmap to visualize this better
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams["figure.figsize"] = (7, 5)  # params are (width_size, height_size)
plt.rcParams["font.size"] = 10

plt.show()
```

```python
# Note: Looking at this heatmap, we can see that the Hours Studied and Attendance has a decent correlation with the Exam Score.
#   So we can infer that it is possible that students who study more and attend more classes have higher exam scores.
```

```python
# let's look at the averages based on Teacher Quality - remember to use numeric only or else it errors out
df.groupby("Teacher_Quality").mean(numeric_only = True).sort_values(by = "Exam_Score", ascending=False)
```

```python
# So looking at the values here, we can infer that the Teacher Quality didn't really have an impact on students, speaking in averages
```

```python
# Let's look at the average of Exam Score based on Hours Studied
df2 = df.groupby("Hours_Studied")["Exam_Score"].mean()
df2
```

```python
# if we wanted to switch axises, use df.transpose() function
df2.plot()
```

```python
# The above line graph shows linear progression between hours studied and exam scores. There are a few outliers where the exam score is not following progression around 0-3 hours studied
# and around 43-50 hours studied. There may be other factors which caused these students to do better or worse than others in their range.
```

```python
# Let's use boxplot to look more into those outliers
df.boxplot(column="Hours_Studied")
```

```python
# might be better to find outliers with scatter plot
df.plot.scatter(x = "Hours_Studied", y = "Exam_Score")
```

```python
# Let's filter those values in the table to see if there might be a reason for the outliers
df.query("Hours_Studied <= 10 & Exam_Score >= 80").sort_values("Exam_Score", ascending=False)
```

```python
# Looking at this query, though the student with the highest score has only 1 hour of study time, 
# he was probably able to get the highest score because he had 3 tutoring sessions over the 0-1 sessions the other students had.
# There might be other factors but I believe that is the most prominent.
```
